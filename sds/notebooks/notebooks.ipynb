```python
import graphviz
import numpy as np

cap_colors = ['pink', 'green', 'red', 'brown']
costume_colors = ['black', 'blue', 'yellow']

tree = 'digraph { layout="neato" bgcolor="#00000000" s[pos="0,0!" label="?"] \n'

cap_rho = 1.3
cap_theta = np.linspace(0, np.pi, len(cap_colors))
cap_pos = np.array((cap_rho * np.cos(cap_theta), cap_rho * np.sin(cap_theta))).T

costume_rho = 1
theta_diff = np.pi/2 /(len(costume_colors)+1)
theta_start = 0
theta_end = 2 * np.pi/6
theta_gap = (len(costume_colors) + 1) / len(costume_colors) * np.pi/6

for cap_color, (x, y) in zip(cap_colors, cap_pos):
    tree += f'cap_{cap_color}[pos="{x:.2f},{y:.2f}!" image="./p3-CP/c1-calcolo-combinatorio/img/{cap_color}-cap.png" label="", style="invisible" width=.6 height=.6 fixedsize="true"]\n'
    tree += f's -> cap_{cap_color};'
    

    
    costume_theta = np.linspace(theta_start, theta_end, len(costume_colors))
    costume_pos = np.array((costume_rho * np.cos(costume_theta), costume_rho * np.sin(costume_theta))).T
    theta_start += theta_gap
    theta_end += theta_gap

    for costume_color, (x_delta, y_delta) in zip(costume_colors, costume_pos):
        tree += f'cap_{cap_color}_costume_{costume_color}[pos="{x+x_delta:.2f},{y+y_delta:.2f}!" image="./p3-CP/c1-calcolo-combinatorio/img/sh-{costume_color}-{cap_color}.png" label="", style="invisible" width=.6 height=.6 fixedsize="true"]\n'
        tree += f'cap_{cap_color} -> cap_{cap_color}_costume_{costume_color};'
    

tree += '}'

graphviz.Source(tree)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

plt.rcParams.update({
    'font.size': 8,
    #'text.usetex': True,
    'text.latex.preamble': r'\usepackage{amsfonts}'
})

# to ensure that pdf is a probability density
Z = norm()
norm_factor = (1 - Z.cdf(3 * 2**0.5))/ 2**0.5 + 5**0.5

def pdf(x):
    return (1.2 * np.exp(-.1*(x)**2) + np.exp(-(x-3)**2)) / norm_factor

fig, ax = plt.subplots()
x = np.linspace(-0, 5, 500)
y = pdf(x)

x_filled = np.linspace(1, 3, 500)
y_filled = pdf(x_filled)

ax.fill_between(x_filled, y_filled, alpha=0.5)
ax.plot(x, y)

ax.text(2, -.1, r'$A$', horizontalalignment='center')
ax.text(2, 0.3, '$P(X \\in A)$', horizontalalignment='center')

ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.tick_params(axis=u'both', which=u'both',length=0)
plt.show()
```

```python
def cont(x, p):
    return (1 - p) ** (x / p)

def lim(x):
    return np.exp(-x)

def disc(x, p):
    return (1 - p) ** (x / p).astype(int)
```

```python
import matplotlib.pyplot as plt

def geom_cdf(x, p):
    return 1 - (1 - p) ** (np.floor(x) + 1)

def expo_cdf(x, lambda_):
    return 1 - np.exp(-lambda_ * x)

x = np.arange(0, 100)

lambda_ = 1

for n in range(2, 5):
    p = lambda_ / n

    y = geom_cdf(x*n, p)

    plt.step(x*n, y, label=n) 

x = np.linspace(0, 10, 100)
y = expo_cdf(x, lambda_)
plt.plot(x, y)
plt.xlim(0, 12)
plt.legend()     
plt.show()
```

```python
import ipywidgets as widgets
from scipy import stats as st

param_slider = widgets.FloatRangeSlider(value=[-3, 3],
                                        min=-10,
                                        max=10,
                                        step=0.5,
                                        description='(a, b)',
                                        continuous_update=True,
                                        readout=False,
                                        orientation='horizontal')

def uniform_kurtosis(params):
    a, b = params
    if(a < b):
        x = np.linspace(-10, 10, 200)
        Z = st.norm((a+b)/2, (b-a)/(12)**0.5)
        y_norm = Z.pdf(x)

        plt.plot(x, y_norm)

        plt.plot([-10, a], [0, 0], c='k')
        plt.plot([a, b], [1/(b-a), 1/(b-a)], c='k')
        plt.plot([b, 10], [0, 0], c='k')
        plt.ylim(0, 0.8)
        plt.title(rf'$(a, b) = ({a:.2f}, {b:.2f})$')
        plt.show()

widgets.interactive(uniform_kurtosis, params = param_slider)
```

```python
import ipywidgets as widgets
from scipy import stats as st

param_slider = widgets.FloatSlider(value=0.5,
                                        min=0.1,
                                        max=0.99,
                                        step=0.01,
                                        description='p',
                                        continuous_update=True,
                                        readout=False,
                                        orientation='horizontal')

def bernoulli_kurtosis(p):
    x = np.linspace(-10, 10, 200)
    Z = st.norm(p, (p*(1 - p))**0.5)
    y_norm = Z.pdf(x)

    plt.plot(x, y_norm)

    plt.plot([0, 0], [0, 1-p], c='k')
    plt.plot([1, 1], [0, p], c='k')
    plt.title(rf'$p = {p:.2f}$')
    plt.show()

widgets.interactive(bernoulli_kurtosis, p = param_slider)
```
